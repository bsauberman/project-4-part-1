```python
import tensorflow as tf # Import tensorflow libraryimport matplotlib.pyplot as plt # data visualization library
from matplotlib import pyplot
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mnist = tf.keras.datasets.mnist # Object of the MNIST dataset
(x_train, y_train),(x_test, y_test) = mnist.load_data() # Load data
```

```python
# plot first few images
for i in range(9):
	# define subplot
	pyplot.subplot(330 + 1 + i)
	# plot raw pixel data
	pyplot.imshow(x_train[i], cmap=pyplot.get_cmap('gray'))
# show the figure
pyplot.show()
```

```python
# Normalize the train dataset
x_train = tf.keras.utils.normalize(x_train, axis=1)
# Normalize the test dataset
x_test = tf.keras.utils.normalize(x_test, axis=1)
```

```python
# Build the model object (FINISH ME)
hiddenUnits = [5, 10, 20, 40, 80]
accuracies = []

for h in hiddenUnits:
    model = tf.keras.models.Sequential()
    # Add the Flatten Layer
    model.add(tf.keras.layers.Flatten())
    # Build the input and the hidden layers using the amount of hiddenUnits specified in array
    model.add(tf.keras.layers.Dense(128, activation="relu"))
    model.add(tf.keras.layers.Dense(h, activation="relu"))
    # Build the output layer
    model.add(tf.keras.layers.Dense(10, activation=tf.nn.softmax))
    
    # Compile the model
    model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

    model.fit(x=x_train, y=y_train, epochs=100) # Start training process with 100 epochs (what we did in part 1(1 ) 1)


    results = model.evaluate(x_test, y_test, verbose = 0)
    
    print("Hidden Units: " + f"{h}{round(results[1], 3)}")
    accuracy.append(results[1])
```

```python
predictions = model.predict([x_test]) # Make prediction
```

```python
print(np.argmax(predictions[1000])) # Print out the number
```

```python
plt.imshow(x_test[1000], cmap="gray") # Import the image
plt.show() # Show the image
```

```python
print(accuracies)
```

```python
import matplotlib.pyplot as plt

# PLOTS NUM EPOCHS VS ACCURACY FOR TRAINING SET
plt.plot(hiddenUnits, accuracies, 
         label='Training')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(loc='lower right')
plt.show()
```
